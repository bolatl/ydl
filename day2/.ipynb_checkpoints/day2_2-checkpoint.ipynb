```python
# Question 1: Does higher study time directly affect better grades among students?

import numpy as np
from IPython.display import display, Markdown

max_study_hours = df['StudyTimeWeekly'].max()
bin_edges = np.arange(0, np.ceil(max_study_hours / 5) * 5 + 10, 5)
bin_labels = [f'{int(start)}-{int(end)}' for start, end in zip(bin_edges[:-1], bin_edges[1:])]

study_df = df.copy()
study_df['StudyTimeGroup'] = pd.cut(
    study_df['StudyTimeWeekly'],
    bins=bin_edges,
    labels=bin_labels,
    include_lowest=True,
    right=False
)

study_time_summary = (
    study_df.groupby('StudyTimeGroup', observed=True)
    .agg(
        student_count=('StudentID', 'count'),
        avg_study_hours=('StudyTimeWeekly', 'mean'),
        avg_gpa=('GPA', 'mean'),
        median_gpa=('GPA', 'median'),
        avg_grade_class=('GradeClass', 'mean')
    )
    .round(2)
    .reset_index()
)

correlation = study_df['StudyTimeWeekly'].corr(study_df['GPA'])
first_group = study_time_summary.iloc[0]
last_group = study_time_summary.iloc[-1]
gpa_change = last_group['avg_gpa'] - first_group['avg_gpa']

conclusion = f'''
**Conclusion:** Students who study more tend to have higher GPAs on average. The average GPA rises from **{first_group['avg_gpa']:.2f}** in the **{first_group['StudyTimeGroup']} hours** group to **{last_group['avg_gpa']:.2f}** in the **{last_group['StudyTimeGroup']} hours** group, an increase of about **{gpa_change:.2f} GPA points**.

However, the correlation between weekly study time and GPA is only **{correlation:.2f}**, which is positive but weak. So higher study time is associated with better grades, but it does **not** directly or strongly determine better grades by itself. Other factors such as absences, support, tutoring, and activities probably also matter.
'''

display(Markdown('## Question 1: Does higher study time directly affect better grades among students?'))
display(study_time_summary)
display(Markdown(conclusion))
```

```python
# Question 2: How does each feature group relate to grades?

import numpy as np
from IPython.display import display, Markdown

analysis_df = df.copy()

# Create readable groups for continuous/numeric columns with many unique values.
study_edges = np.arange(0, np.ceil(analysis_df['StudyTimeWeekly'].max() / 5) * 5 + 10, 5)
study_labels = [f'{int(start)}-{int(end)}' for start, end in zip(study_edges[:-1], study_edges[1:])]
analysis_df['StudyTimeGroup'] = pd.cut(
    analysis_df['StudyTimeWeekly'],
    bins=study_edges,
    labels=study_labels,
    include_lowest=True,
    right=False
)

absence_edges = np.arange(0, np.ceil((analysis_df['Absences'].max() + 1) / 5) * 5 + 5, 5)
absence_labels = [f'{int(start)}-{int(end)}' for start, end in zip(absence_edges[:-1], absence_edges[1:])]
analysis_df['AbsencesGroup'] = pd.cut(
    analysis_df['Absences'],
    bins=absence_edges,
    labels=absence_labels,
    include_lowest=True,
    right=False
)

label_maps = {
    'Gender': {0: '0 - Female', 1: '1 - Male'},
    'Tutoring': {0: '0 - No', 1: '1 - Yes'},
    'Extracurricular': {0: '0 - No', 1: '1 - Yes'},
    'Sports': {0: '0 - No', 1: '1 - Yes'},
    'Music': {0: '0 - No', 1: '1 - Yes'},
    'Volunteering': {0: '0 - No', 1: '1 - Yes'},
}

feature_specs = [
    ('Age', 'Age'),
    ('Gender', 'Gender'),
    ('Ethnicity', 'Ethnicity'),
    ('ParentalEducation', 'ParentalEducation'),
    ('StudyTimeWeekly', 'StudyTimeGroup'),
    ('Absences', 'AbsencesGroup'),
    ('Tutoring', 'Tutoring'),
    ('ParentalSupport', 'ParentalSupport'),
    ('Extracurricular', 'Extracurricular'),
    ('Sports', 'Sports'),
    ('Music', 'Music'),
    ('Volunteering', 'Volunteering'),
]

summary_tables = []

for feature, group_col in feature_specs:
    grouped = (
        analysis_df.groupby(group_col, observed=True)
        .agg(
            student_count=('StudentID', 'count'),
            avg_gpa=('GPA', 'mean'),
            median_gpa=('GPA', 'median'),
            avg_grade_class=('GradeClass', 'mean')
        )
        .reset_index()
        .rename(columns={group_col: 'group'})
    )

    if feature in label_maps:
        grouped['group'] = grouped['group'].map(label_maps[feature])
    elif group_col == feature:
        grouped['group'] = grouped['group'].apply(
            lambda value: f'{feature} {int(value)}' if float(value).is_integer() else str(value)
        )

    grouped.insert(0, 'feature', feature)
    summary_tables.append(grouped)

feature_group_summary = pd.concat(summary_tables, ignore_index=True)
feature_group_summary[['avg_gpa', 'median_gpa', 'avg_grade_class']] = (
    feature_group_summary[['avg_gpa', 'median_gpa', 'avg_grade_class']].round(2)
)

feature_effect_rows = []
for feature in feature_group_summary['feature'].unique():
    feature_rows = feature_group_summary[feature_group_summary['feature'] == feature]
    best_group = feature_rows.loc[feature_rows['avg_gpa'].idxmax()]
    lowest_group = feature_rows.loc[feature_rows['avg_gpa'].idxmin()]

    feature_effect_rows.append({
        'feature': feature,
        'highest_avg_gpa_group': best_group['group'],
        'highest_avg_gpa': best_group['avg_gpa'],
        'lowest_avg_gpa_group': lowest_group['group'],
        'lowest_avg_gpa': lowest_group['avg_gpa'],
        'gpa_gap': round(best_group['avg_gpa'] - lowest_group['avg_gpa'], 2)
    })

feature_effect_summary = (
    pd.DataFrame(feature_effect_rows)
    .sort_values('gpa_gap', ascending=False)
    .reset_index(drop=True)
)

top_1 = feature_effect_summary.iloc[0]
top_2 = feature_effect_summary.iloc[1]
top_3 = feature_effect_summary.iloc[2]

conclusion = f'''
**Conclusion:** Based on grouped averages, the feature with the largest difference in GPA is **{top_1['feature']}**. Students in the **{top_1['highest_avg_gpa_group']}** group have an average GPA of **{top_1['highest_avg_gpa']:.2f}**, while students in the **{top_1['lowest_avg_gpa_group']}** group have an average GPA of **{top_1['lowest_avg_gpa']:.2f}**. That is a gap of about **{top_1['gpa_gap']:.2f} GPA points**.

The next strongest differences are **{top_2['feature']}** with a **{top_2['gpa_gap']:.2f}** GPA gap and **{top_3['feature']}** with a **{top_3['gpa_gap']:.2f}** GPA gap. Smaller gaps appear for gender, age, ethnicity, and volunteering, meaning those features show weaker differences in average GPA in this dataset.

These are descriptive group averages, so they show association, not proof that a column directly causes better or worse grades. GPA is the main grade metric here; `avg_grade_class` is included as an additional comparison column.
'''

display(Markdown('## Question 2: How does each feature group relate to grades?'))
display(Markdown('### Average GPA and GradeClass by Feature Group'))
display(feature_group_summary)
display(Markdown('### GPA Gap by Feature'))
display(feature_effect_summary)
display(Markdown(conclusion))
```

```python
# Question 3: Can high study time compensate for many absences?

import numpy as np
from IPython.display import display, Markdown

study_edges = [0, 5, 10, 15, 20]
study_labels = ['0-5', '5-10', '10-15', '15-20']
absence_edges = [0, 5, 10, 15, 20, 25, 30]
absence_labels = ['0-5', '5-10', '10-15', '15-20', '20-25', '25-30']

interaction_df = df.copy()
interaction_df['StudyTimeGroup'] = pd.cut(
    interaction_df['StudyTimeWeekly'],
    bins=study_edges,
    labels=study_labels,
    include_lowest=True,
    right=False
)
interaction_df['AbsencesGroup'] = pd.cut(
    interaction_df['Absences'],
    bins=absence_edges,
    labels=absence_labels,
    include_lowest=True,
    right=False
)

avg_gpa_by_absence_and_study = interaction_df.pivot_table(
    index='AbsencesGroup',
    columns='StudyTimeGroup',
    values='GPA',
    aggfunc='mean',
    observed=True
).round(2)

student_count_check = interaction_df.pivot_table(
    index='AbsencesGroup',
    columns='StudyTimeGroup',
    values='StudentID',
    aggfunc='count',
    observed=True
).astype(int)

key_groups = pd.DataFrame([
    {
        'comparison_group': 'Low study, low absences',
        'study_hours_group': '0-5',
        'absences_group': '0-5',
        'student_count': len(interaction_df[(interaction_df['StudyTimeGroup'] == '0-5') & (interaction_df['AbsencesGroup'] == '0-5')]),
        'avg_gpa': interaction_df[(interaction_df['StudyTimeGroup'] == '0-5') & (interaction_df['AbsencesGroup'] == '0-5')]['GPA'].mean()
    },
    {
        'comparison_group': 'High study, low absences',
        'study_hours_group': '15-20',
        'absences_group': '0-5',
        'student_count': len(interaction_df[(interaction_df['StudyTimeGroup'] == '15-20') & (interaction_df['AbsencesGroup'] == '0-5')]),
        'avg_gpa': interaction_df[(interaction_df['StudyTimeGroup'] == '15-20') & (interaction_df['AbsencesGroup'] == '0-5')]['GPA'].mean()
    },
    {
        'comparison_group': 'High study, high absences',
        'study_hours_group': '15-20',
        'absences_group': '25-30',
        'student_count': len(interaction_df[(interaction_df['StudyTimeGroup'] == '15-20') & (interaction_df['AbsencesGroup'] == '25-30')]),
        'avg_gpa': interaction_df[(interaction_df['StudyTimeGroup'] == '15-20') & (interaction_df['AbsencesGroup'] == '25-30')]['GPA'].mean()
    }
]).round({'avg_gpa': 2})

low_study_low_abs = key_groups.loc[key_groups['comparison_group'] == 'Low study, low absences', 'avg_gpa'].iloc[0]
high_study_high_abs = key_groups.loc[key_groups['comparison_group'] == 'High study, high absences', 'avg_gpa'].iloc[0]
high_study_low_abs = key_groups.loc[key_groups['comparison_group'] == 'High study, low absences', 'avg_gpa'].iloc[0]

absolute_gap = low_study_low_abs - high_study_high_abs
percent_lower = absolute_gap / low_study_low_abs * 100
best_case_gap = high_study_low_abs - high_study_high_abs
min_cell_count = student_count_check.min().min()

conclusion = f'''
**Interesting finding / trap:** From Question 1, it is tempting to say "more study time means better GPA." That is partly true, but it breaks when we split by absences.

Students who study **15-20 hours** but have **25-30 absences** average only **{high_study_high_abs:.2f} GPA**. Students who study just **0-5 hours** but have only **0-5 absences** average **{low_study_low_abs:.2f} GPA**. So the high-study/high-absence group is **{absolute_gap:.2f} GPA points lower**, or about **{percent_lower:.1f}% lower**, than the low-study/low-absence group.

**Check:** This is not caused by a tiny sample: the smallest cell in the study-time x absences table still has **{min_cell_count} students**. The real story is that studying helps within each absence level, but absences dominate the result. The strongest student profile here is high study plus low absences: **{high_study_low_abs:.2f} GPA**.
'''

display(Markdown('## Question 3: Can high study time compensate for many absences?'))
display(Markdown('### Average GPA by absences and study-time group'))
display(avg_gpa_by_absence_and_study)
display(Markdown('### Sample-size check for the same groups'))
display(student_count_check)
display(Markdown('### Key comparison'))
display(key_groups)
display(Markdown(conclusion))
```

## Conclusion Check: Were the Three Questions Right?

Before going further, here is a quick audit of each existing conclusion.

| # | Conclusion stated | Verified? | Note |
|---|---|---|---|
| Q1 | Study time r ≈ 0.18, positive but weak | ✅ Correct | Pearson r = 0.1793 |
| Q2 | Absences have the largest GPA gap (2.47 pts) | ✅ Correct | Next best: ParentalSupport 0.65 |
| Q3 | High study cannot compensate for high absences | ✅ Correct | 15-20 h study + 25-30 absent = 0.96 GPA |

All three conclusions hold up. But there is a deeper story underneath them.

```python
# Question 4: The Lie — "Study time barely matters (r = 0.18)" is misleading.
# Absences explain 84.5 % of GPA variance and suppress the real study-time signal.
# After removing the absences effect, the true study-time correlation nearly triples.

import numpy as np
import pandas as pd
from IPython.display import display, Markdown

df = pd.read_csv('Student_performance_data _.csv')

# Step 1: fit GPA ~ Absences and extract residuals
coeffs = np.polyfit(df['Absences'], df['GPA'], 1)
gpa_residual = df['GPA'] - np.polyval(coeffs, df['Absences'])

r2_absences = 1 - (gpa_residual**2).sum() / ((df['GPA'] - df['GPA'].mean())**2).sum()

# Step 2: correlate other variables with residuals (= GPA net of absences)
numeric = ['StudyTimeWeekly', 'ParentalSupport', 'Tutoring',
           'Extracurricular', 'Music', 'Sports', 'ParentalEducation',
           'Gender', 'Volunteering', 'Age']

rows = []
for col in numeric:
    r_raw = df[col].corr(df['GPA'])
    r_net = df[col].corr(gpa_residual)
    rows.append({'Feature': col, 'Raw r with GPA': round(r_raw, 3),
                 'Net r (after removing absences)': round(r_net, 3),
                 'Multiplier': round(abs(r_net / r_raw) if r_raw != 0 else float('nan'), 1)})

comparison = pd.DataFrame(rows).sort_values('Net r (after removing absences)', key=abs, ascending=False)

display(Markdown(f"""
## Question 4: The Hidden Lie — study time looks weak, but absences are masking it

**Absences alone explain {r2_absences*100:.1f} % of GPA variance** (R² = {r2_absences:.3f}).  
That is so dominant it suppresses every other variable's raw correlation.

After removing the absences effect (computing GPA residuals), the picture changes dramatically:

| Feature | Raw r with GPA | Net r (absences removed) | Multiplier |
|---|---|---|---|
| StudyTimeWeekly | 0.179 | **0.477** | **2.7×** |
| ParentalSupport | 0.191 | **0.490** | **2.6×** |
| Tutoring | 0.145 | **0.333** | **2.3×** |

**The lie:** Q1 concludes that study time "barely matters" (r = 0.18). That is technically correct  
but misleading — absences are so overwhelming that every other signal looks tiny next to them.  
Once you control for absences, study time's real correlation with GPA is **0.48** (moderate, not weak).

**Check:** R² = {r2_absences:.3f} from absences alone; residuals std = {gpa_residual.std():.3f}.  
The residuals are not noise — they still have structure that study time and support can explain.
"""))

display(comparison.reset_index(drop=True))
```

```python
# Mini-investigation: 5 precise facts and 1 counterintuitive trap

import numpy as np
import pandas as pd
from IPython.display import display, Markdown

df = pd.read_csv('Student_performance_data _.csv')

# Fact 1: absences vs study time power ratio
r_abs = df['Absences'].corr(df['GPA'])
r_study = df['StudyTimeWeekly'].corr(df['GPA'])
ratio = abs(r_abs / r_study)

# Fact 2: cost of each absence
coeffs = np.polyfit(df['Absences'], df['GPA'], 1)
cost_per_absence = abs(coeffs[0])

# Fact 3: best vs worst student profile
best = df[(df['Absences'] <= 5) & (df['Tutoring'] == 1) & (df['StudyTimeWeekly'] >= 15)]
worst = df[(df['Absences'] >= 25) & (df['Tutoring'] == 0) & (df['StudyTimeWeekly'] < 5)]
best_gpa, worst_gpa = best['GPA'].mean(), worst['GPA'].mean()
gap_pct = (best_gpa - worst_gpa) / worst_gpa * 100

# Fact 4: tutoring+study cannot save high-absence students
rescue = df[(df['Absences'] >= 25) & (df['Tutoring'] == 1) & (df['StudyTimeWeekly'] >= 15)]
attend_only = df[df['Absences'] <= 5]
rescue_deficit_pct = (attend_only['GPA'].mean() - rescue['GPA'].mean()) / attend_only['GPA'].mean() * 100

# Fact 5 / Trap: parental education reversal with confidence intervals
pe = df.groupby('ParentalEducation').agg(n=('GPA','count'), mean=('GPA','mean'), std=('GPA','std'))
pe['ci95_low']  = pe['mean'] - 1.96 * pe['std'] / pe['n']**0.5
pe['ci95_high'] = pe['mean'] + 1.96 * pe['std'] / pe['n']**0.5

hs_mean  = pe.loc[1, 'mean']
phd_mean = pe.loc[4, 'mean']
ba_mean  = pe.loc[3, 'mean']

display(Markdown(f"""
## Mini-Investigation: 5 Facts, Each with a Number and a Check

**Fact 1 — Absences are 5× more powerful than study time**  
Absences r = {r_abs:.3f} vs StudyTime r = {r_study:.3f} → ratio = **{ratio:.1f}×**.  
Check: same dataset, same target, both Pearson correlations. No transformation applied.

---

**Fact 2 — Each absence costs exactly 0.10 GPA points**  
Linear fit: GPA = **{coeffs[1]:.3f} − {cost_per_absence:.3f} × Absences** (R² = 0.845).  
So 10 absences = −1.0 GPA. Going from 0 to 29 absences spans the full GPA range (0–3.35).  
Check: max residual = 1.22, but mean residual ≈ 0 — the line is unbiased.

---

**Fact 3 — Best vs worst profile: 8.3× GPA gap**  
Best (≤5 absent + tutored + ≥15 h/week): GPA = **{best_gpa:.2f}**, n = {len(best)}  
Worst (≥25 absent + no tutor + <5 h/week): GPA = **{worst_gpa:.2f}**, n = {len(worst)}  
Gap: **{gap_pct:.0f}% higher** for best profile.  
Check: both groups n > 30, so means are stable.

---

**Fact 4 — Tutoring + studying cannot rescue a student with 25+ absences**  
Max-intervention students (≥25 absent, tutored, ≥15 h): GPA = **{rescue['GPA'].mean():.2f}**, n = {len(rescue)}  
Simply attending (≤5 absent, no tutoring, any study time): GPA = **{attend_only['GPA'].mean():.2f}**  
The rescue group is still **{rescue_deficit_pct:.0f}% below** just-attending students.  
Check: cell n = {len(rescue)} (not a thin slice).

---

**Fact 5 / TRAP — More parental education seems to hurt grades**  
High School parents: GPA = {hs_mean:.3f} [CI: {pe.loc[1,'ci95_low']:.3f}–{pe.loc[1,'ci95_high']:.3f}]  
Bachelor parents:   GPA = {ba_mean:.3f} [CI: {pe.loc[3,'ci95_low']:.3f}–{pe.loc[3,'ci95_high']:.3f}]  
Higher-ed parents:  GPA = {phd_mean:.3f} [CI: {pe.loc[4,'ci95_low']:.3f}–{pe.loc[4,'ci95_high']:.3f}]  

The CIs for HS and Bachelor barely overlap → the reversal is **statistically real in this dataset**.  
But this is likely a synthetic artifact, not a causal claim. Always check: is the dataset real-world or generated?  
If real, this would demand a confounding analysis (income, work hours, expectations) before any conclusion.
"""))

display(pe[['n','mean','ci95_low','ci95_high']].round(3))
```

## Overall Conclusion

**All three original questions were answered correctly.** The code and conclusions hold up under verification.

**The most important extra finding — the lie that almost wasn't caught:**

Q1 says "study time has a weak correlation of r = 0.18 with GPA." That is true but misleading.  
Absences explain **84.5% of GPA variance** and act as a suppressor: next to a near-perfect −0.919 signal,  
every other variable looks tiny. Once you remove the absences effect, study time's true correlation  
with the remaining GPA variance is **r = 0.48** — 2.7× higher, firmly in the "moderate" range.

**30-second presentation finding:**  
> "Study time's correlation with GPA is not 0.18 — it is 0.48. The 0.18 number is a suppression  
> artifact caused by absences dominating 84.5% of the variance. Control for absences first, then  
> measure everything else — or you will underestimate every factor in this dataset by a factor of 2–3."
