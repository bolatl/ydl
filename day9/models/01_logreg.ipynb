# 01 - Logistic Regression (linear baseline)

**Idea.** Logistic regression draws *straight* (linear) decision boundaries between the 4 sleep stages and outputs a probability per class.

**Note on "linear regression".** Plain linear regression predicts a continuous number and is the wrong tool for unordered categories (it would imply stage 3 is "3x" stage 1 and can't give class probabilities). Logistic regression is the correct linear classifier, so we use it here.

**Good for this data:** fast, interpretable (coefficients show which signals push toward which stage), solid sanity-check baseline, useful ensemble member (its errors differ from trees').

**Bad for this data:** can only draw straight boundaries, but the classes are nonlinearly separable here, so it has a low ceiling (~0.745). Needs feature scaling and can't ingest NaNs directly.

**Expected CV macro-F1:** ~0.745

```python
import warnings; warnings.filterwarnings('ignore')
import numpy as np, pandas as pd

# Notebooks live in day9/models/, data sits one level up in day9/
TRAIN_PATH = '../train.csv'
TEST_PATH  = '../test.csv'
OUT_DIR    = '../outputs'

train = pd.read_csv(TRAIN_PATH)
test  = pd.read_csv(TEST_PATH)

# every column except the id and the label is a predictor
FEATURES = [c for c in train.columns if c not in ('id', 'sleep_stage')]
TARGET   = 'sleep_stage'
y = train[TARGET]
print('train', train.shape, '| test', test.shape, '| #features', len(FEATURES))
train.head()
```

## Preprocessing
Two things this model needs that tree models don't:
1. **Median imputation** - `eog_burst_index` is ~50% NaN and logistic regression cannot accept NaN.
2. **StandardScaler** - features have different spreads (std 1-5); scaling puts them on equal footing so the regularization treats them fairly.
We wrap both in a `Pipeline` so they are re-fit *inside* each CV fold (no leakage).

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression

# impute -> scale -> logistic regression, as one estimator
clf = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler(),
    LogisticRegression(max_iter=2000),
)
clf
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold

# SAME validation protocol for every model so scores are comparable:
#   - 5 folds, stratified  -> each fold keeps the 22-27% class balance
#   - shuffle + fixed seed  -> reproducible split
#   - scoring = macro-F1    -> the competition metric (every class weighted equally)
# cross_val_score trains on 4 folds and scores the held-out fold, 5x, and averages.
cv = StratifiedKFold(5, shuffle=True, random_state=42)

def benchmark(model, X, label):
    s = cross_val_score(model, X, y, cv=cv, scoring='f1_macro')
    print(f'{label}: macro-F1 = {s.mean():.4f} +/- {s.std():.4f}')
    return s
```

## Benchmark

```python
benchmark(clf, train[FEATURES], 'LogReg')
```

## Create submission
We retrain the same pipeline on all training rows and predict the test set. The pipeline imputes+scales the test data using statistics learned from train.

```python
import os
os.makedirs(OUT_DIR, exist_ok=True)

# LogReg needs imputation + scaling, handled inside the pipeline.
final_model = make_pipeline(SimpleImputer(strategy='median'), StandardScaler(), LogisticRegression(max_iter=2000))
final_model.fit(train[FEATURES], y)            # train on ALL 9000 labelled rows
pred = final_model.predict(test[FEATURES])     # predict the 5000 unlabelled test rows

submission = pd.DataFrame({'id': test['id'], 'sleep_stage': np.asarray(pred).astype(int)})
path = os.path.join(OUT_DIR, 'logreg_submission.csv')
submission.to_csv(path, index=False)
print('wrote', path, submission.shape)
submission.head()
```

## How to improve later
- Tune `C` (regularization) with `GridSearchCV`.
- Add polynomial / interaction features (`PolynomialFeatures`) to fake some nonlinearity.
- Drop one of each highly-correlated pair (corr up to 0.82) to stabilize coefficients.
- Keep it as a member in a final ensemble.
